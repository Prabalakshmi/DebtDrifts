```python
import pandas as pd
```

```python
df_studentloanforgiveness = pd.read_csv('student_loan_forgiveness.csv')
df_studentdebtbomb = pd.read_csv('student_debt_bomb.csv')
df_loanforgiveness = pd.read_csv('loan_forgiveness.csv')
df_studentloandebt = pd.read_csv('student_loan_debt.csv')
df_studentdebt = pd.read_csv('student_debt.csv')
df_studentloan = pd.read_csv('student_loan.csv')
```

```python
frames = [df_studentloanforgiveness, df_studentdebtbomb, df_loanforgiveness, df_studentloandebt, df_studentdebt, df_studentloan]
  
df = pd.concat(frames)

display(df)

print(len(df))
```

```python
df['Location'] = 'United States'
df.drop(['User Location'], axis=1, inplace=True)
df.head(5)
```

```python
# Data Cleaning

# Using regex to clean the data

import re
import nltk
from nltk.corpus import stopwords

import numpy as np
import re

#df = pd.read_csv('fff.csv')

stopwords = set(stopwords.words('english'))

# stopwords.add('amp', 'AMP', '&amp', '&', '&AMP')
stopwords.add('amp')
stopwords.add('AMP')
stopwords.add('&')
stopwords.add('&amp')
stopwords.add('&AMP')


def clean_tweet(tweet):
    if type(tweet) == np.float:
        return ""
    temp = tweet.lower()
    temp = re.sub("'", "", temp) # to avoid removing contractions in english
    temp = re.sub("@[A-Za-z0-9_]+","", temp)
    temp = re.sub("#[A-Za-z0-9_]+","", temp)
    temp = re.sub(r'http\S+', '', temp)
    temp = re.sub('[()!?]', ' ', temp)
    temp = re.sub('\[.*?\]',' ', temp)
    #temp = re.sub("[^a-z0-9]"," ", temp) --> Doesn't remove numbers
    temp = re.sub("[^a-z]"," ", temp)    
    temp = temp.split()
    temp = [w for w in temp if not w in stopwords]
    temp = " ".join(word for word in temp)
    return temp
```

```python
results = [clean_tweet(tw) for tw in df['Tweet']]
results[:10]
#df.head(10)
```

```python
#create new df from list

df['Cleaned_Tweet'] = results
df.head(5)
```

```python
# Dropping rows that have empty values as tweets after data cleaning because without dropping those I can't use it in
# sentimentintensityanalyser() function, as it only accepts string input and empty values are of float type

print('Number of Tweets before data cleaning:', df.shape[0])

df.dropna(
    axis=0,
    how='any',
    thresh=None,
    subset=None,
    inplace=True
)

print('Number of Tweets after data cleaning:', df.shape[0])
#print(df['Cleaned_Tweet'][3])
df.head(3)
```

```python
df.to_csv('Cleaned_Twitter_Data.csv')
```
